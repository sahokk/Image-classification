```python
import cv2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import math
import sys
from google.colab import files
from google.colab.patches import cv2_imshow
```

```python
uploaded = files.upload()
img = cv2.imread('skin1.jpg', cv2.IMREAD_UNCHANGED)
img2 = cv2.imread('skin2.jpg', cv2.IMREAD_UNCHANGED)
```

```python
re_img = cv2.resize(img,(200,200))
re_img2 = cv2.resize(img2,(200,200))
fig = plt.figure(dpi=100, figsize=(8,8))

cv2_imshow(re_img)
cv2_imshow(re_img2)
```

```python
gray_img = cv2.cvtColor(re_img, cv2.COLOR_BGR2GRAY)
gray_img2 = cv2.cvtColor(re_img2, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_img)
cv2_imshow(gray_img2)
```

```python
#
ret, binary = cv2.threshold(gray_img2,180,250,cv2.THRESH_BINARY)

#
plt.imshow(binary, cmap='gray')
```

```python
canny_img = cv2.Canny(gray_img2, 50, 90)

ax = fig.add_subplot(1, 2, 2)
ax.set_title("canny",fontsize=20)
plt.imshow(canny_img, cmap='gray')
```
