## color 32*32


```python
im_size = 32
num_classes = 3
im_color = 3
im_shape = (im_size, im_size, im_color)
```

```python
from sklearn.model_selection import train_test_split
import random
import os
from google.colab import drive
drive.mount('./gdrive')
random.seed(43)

hamster = "./gdrive/My Drive/2022年_3年/1_深層学習_TK200519_2022/期末課題/dataset_color_32/hamster"
guineapig = "./gdrive/My Drive/2022年_3年/1_深層学習_TK200519_2022/期末課題/dataset_color_32/guinea pig"
rabbit = "./gdrive/My Drive/2022年_3年/1_深層学習_TK200519_2022/期末課題/dataset_color_32/rabbit"

hamster_fnames = os.listdir( hamster )
guineapig_fnames = os.listdir( guineapig )
rabbit_fnames = os.listdir( rabbit )

print(hamster_fnames[:5])
print(guineapig_fnames[:5])
print(rabbit_fnames[:5])
```

```python
#訓練データ、テストデータの作成
from PIL import Image  #画像データ取り扱いモジュール
import numpy as np
from sklearn.model_selection import train_test_split
from tensorflow.keras import utils as np_utils

X = []
y = []
for fname in hamster_fnames:
    im = Image.open(os.path.join(hamster,fname))
    im = np.array(im)
    im = im.astype('float32') / 255
    X.append(im)
    y.append(0)
for fname in guineapig_fnames:
    im = Image.open(os.path.join(guineapig,fname))
    im = np.array(im)
    im = im.astype('float32') / 255
    X.append(im)
    y.append(1)
for fname in rabbit_fnames:
    im = Image.open(os.path.join(rabbit,fname))
    im = np.array(im)
    im = im.astype('float32') / 255
    X.append(im)
    y.append(2)

X = np.array(X)
y = np.array(y)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=100, stratify=y)
y_train = np_utils.to_categorical(y_train, 3)
y_test = np_utils.to_categorical(y_test, 3)
```

```python
import tensorflow as tf
tf.keras.backend.clear_session()
tf.random.set_seed(123)
np.random.seed(35)
labels = ['ハムスター','モルモット','うさぎ']

model = tf.keras.models.Sequential([
  tf.keras.layers.Conv2D(32, (3, 3), padding='same', input_shape=im_shape),
  tf.keras.layers.Activation('relu'),
  tf.keras.layers.Conv2D(32, (3, 3), padding='same'),
  tf.keras.layers.Activation('relu'),
  tf.keras.layers.MaxPooling2D(pool_size=(2, 2)),
  tf.keras.layers.Dropout(0.25),
  tf.keras.layers.Conv2D(64, (3, 3), padding='same'),
  tf.keras.layers.Activation('relu'),
  tf.keras.layers.Conv2D(64, (3, 3), padding='same'),
  tf.keras.layers.Activation('relu'),
  tf.keras.layers.MaxPooling2D(pool_size=(2, 2)),
  tf.keras.layers.Dropout(0.25),
  tf.keras.layers.Flatten(),
  tf.keras.layers.Dense(512),
  tf.keras.layers.Activation('relu'),
  tf.keras.layers.Dropout(0.5),
  tf.keras.layers.Dense(num_classes),
  tf.keras.layers.Activation('softmax')
])
model.summary()
```

```python
# modelのコンパイル
model.compile(loss="categorical_crossentropy",optimizer="adam",metrics=["acc"])  #adam optimizer
```

```python
callbacks_list = [
                  tf.keras.callbacks.ModelCheckpoint(filepath="cnn_da_do", save_best_only=True, monitor="val_loss")  # 一番よいモデルを保存
                  ]
```

```python
#学習
%%time
history = model.fit(X_train, y_train, epochs=10, batch_size=32, validation_data=(X_test, y_test), verbose=1, callbacks=callbacks_list)
```

```python
import matplotlib.pyplot as plt

# モデルを評価
score = model.evaluate(X_test, y_test, verbose=1)
print('正解率=', score[1], 'loss=', score[0])

# 学習の様子をグラフへ描画 --- (*5)
# 正解率の推移をプロット
plt.plot(history.history['acc'])
plt.plot(history.history['val_acc'])
plt.title('Accuracy')
plt.legend(['train', 'test'], loc='upper left')
plt.show()

# ロスの推移をプロット
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Loss')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

```python
from matplotlib import pyplot
probability_model = tf.keras.Sequential([
  model,
  tf.keras.layers.Softmax()
])
tests = X_test[50:58]
pyplot.figure(figsize=(10,10))
for i in range(0, len(tests)):
    pyplot.subplot(1, len(tests), i + 1)
    pyplot.imshow(tests[i], cmap='gray')
pyplot.show()
#tests = tests.reshape(tests.shape[0], 28, 28, 1).astype('float32')
r = probability_model(tests)
result = []
for i in np.argmax(r.numpy(), axis=1):
    result.append(labels[i])
print(result)
```

```python
im_size = 64
num_classes = 3
im_color = 3
im_shape = (im_size, im_size, im_color)
```

```python
from sklearn.model_selection import train_test_split
import random
import os
from google.colab import drive
drive.mount('./gdrive')
random.seed(43)

hamster = "./gdrive/My Drive/2022年_3年/1_深層学習_TK200519_2022/期末課題/dataset_color_64/hamster"
guineapig = "./gdrive/My Drive/2022年_3年/1_深層学習_TK200519_2022/期末課題/dataset_color_64/guinea pig"
rabbit = "./gdrive/My Drive/2022年_3年/1_深層学習_TK200519_2022/期末課題/dataset_color_64/rabbit"

hamster_fnames = os.listdir( hamster )
guineapig_fnames = os.listdir( guineapig )
rabbit_fnames = os.listdir( rabbit )

print(hamster_fnames[:10])
print(guineapig_fnames[:10])
print(rabbit_fnames[:10])
```

```python
#訓練データ、テストデータの作成
from PIL import Image  #画像データ取り扱いモジュール
import numpy as np
from sklearn.model_selection import train_test_split
from tensorflow.keras import utils as np_utils

X = []
y = []
for fname in hamster_fnames:
    im = Image.open(os.path.join(hamster,fname))
    im = np.array(im)
    im = im.astype('float32') / 255
    X.append(im)
    y.append(0)
for fname in guineapig_fnames:
    im = Image.open(os.path.join(guineapig,fname))
    im = np.array(im)
    im = im.astype('float32') / 255
    X.append(im)
    y.append(1)
for fname in rabbit_fnames:
    im = Image.open(os.path.join(rabbit,fname))
    im = np.array(im)
    im = im.astype('float32') / 255
    X.append(im)
    y.append(2)

X = np.array(X)
y = np.array(y)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=100, stratify=y)
y_train = np_utils.to_categorical(y_train, 3)
y_test = np_utils.to_categorical(y_test, 3)
```

```python
y_train[1]
```

```python
from PIL import Image
from matplotlib import pyplot
pyplot.figure(figsize=(10,10))
for i in range(0, 40):
    im = Image.fromarray((X_train[i] * 255).astype(np.uint8))
    pyplot.subplot(5, 8, i + 1)
    pyplot.imshow(im)
pyplot.show()
```

```python
import tensorflow as tf
tf.keras.backend.clear_session()
tf.random.set_seed(123)
np.random.seed(35)
labels = ['ハムスター','モルモット','うさぎ']

model = tf.keras.models.Sequential([
  tf.keras.layers.Conv2D(32, (3, 3), padding='same', input_shape=im_shape),
  tf.keras.layers.Activation('relu'),
  tf.keras.layers.Conv2D(32, (3, 3), padding='same'),
  tf.keras.layers.Activation('relu'),
  tf.keras.layers.MaxPooling2D(pool_size=(2, 2)),
  tf.keras.layers.Dropout(0.25),
  tf.keras.layers.Conv2D(64, (3, 3), padding='same'),
  tf.keras.layers.Activation('relu'),
  tf.keras.layers.Conv2D(64, (3, 3), padding='same'),
  tf.keras.layers.Activation('relu'),
  tf.keras.layers.MaxPooling2D(pool_size=(2, 2)),
  tf.keras.layers.Dropout(0.25),
  tf.keras.layers.Flatten(),
  tf.keras.layers.Dense(512),
  tf.keras.layers.Activation('relu'),
  tf.keras.layers.Dropout(0.5),
  tf.keras.layers.Dense(num_classes),
  tf.keras.layers.Activation('softmax')
])
model.summary()
```

```python
# modelのコンパイル
model.compile(loss="categorical_crossentropy",optimizer="adam",metrics=["acc"])  #adam optimizer
```

```python
callbacks_list = [
                  tf.keras.callbacks.ModelCheckpoint(filepath="cnn_da_do", save_best_only=True, monitor="val_loss")  # 一番よいモデルを保存
                  ]
```

```python
#学習
%%time
history = model.fit(X_train, y_train, epochs=12, batch_size=32, validation_data=(X_test, y_test), verbose=1, callbacks=callbacks_list)
```

```python
import matplotlib.pyplot as plt
# モデルを評価
score = model.evaluate(X_test, y_test, verbose=1)
print('正解率=', score[1], 'loss=', score[0])

# 学習の様子をグラフへ描画 --- (*5)
# 正解率の推移をプロット
plt.plot(history.history['acc'])
plt.plot(history.history['val_acc'])
plt.title('Accuracy')
plt.legend(['train', 'test'], loc='upper left')
plt.show()

# ロスの推移をプロット
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Loss')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```
